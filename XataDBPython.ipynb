```python
import os
from xata.client import XataClient
from xata.helpers import BulkProcessor
from dotenv import load_dotenv
import random
from faker import Faker  
import unittest


load_dotenv()


api_key = os.getenv("XATA_API_KEY")
if api_key is None:
    raise Exception("")

xata = XataClient(api_key=api_key, db_url="https://Gabriel-Moura-s-workspace-a5ttv7.us-east-1.xata.sh/db/BDPJT1:main")
bp = BulkProcessor(xata)
fake = Faker()



descriptions_list = [
    "Audierunt ab Ipiranga ad ripas placidas",
    "Populi heroici, fortis, recens nati.",
    "Et sol libertatis, radiis fulgidis",
    "Splenduit in caelo patriae in hoc momento.",
    "Si vinculum aequalitatis",
    "Vicimus cum brachio forti",
    "In sinu tuo, o libertas",
    "Provocavit cor nostrum ipsam mortem!",
    "O patria amata, adorata, salve! salve!",
    "Brasil, somnium intensum, fulgor vividus",
    "Amoris et spei ad terram descendit.",
    "Si futurum est gloria, praeteritum suum",
    "Patria heroica gloriam a conquestu.",
    "Sol libertatis splendet in hoc caelo immenso",
    "Patria amata, salve! salve!"
]

status_list = ["active", "inactive", "pending"]

num_records = 5000

data = [
    {
        "name": fake.name(),
        "discription": random.choice(descriptions_list),
        "status": random.choice(status_list),
        "email": fake.email(),
        "idade": random.randint(1, 99),
        "address": fake.address()
    }
    for _ in range(num_records)
]

bp.put_records("resources", data)

bp.flush_queue()
```

```python
def fetch_paginated_data(page_size=10):
   
    records = xata.data().query("resources", {
        "page": {
            "size": page_size 
        }
    })

    while True:
        
        if 'records' in records:
            print(f"View {len(records['records'])} ")
            for record in records['records']:
                print(record)

        if records.has_more_results():            
            next_cursor = records.get_cursor()          
            records = xata.data().query("resources", {
                "page": {
                    "after": next_cursor  
                }
            })
        else:
            break
fetch_paginated_data(10)
```

```python
import unittest

def get_translated_hino():
    return [
        "Audierunt ab Ipiranga ad ripas placidas",
        "Populi heroici, fortis, recens nati.",
        "Et sol libertatis, radiis fulgidis",
        "Splenduit in caelo patriae in hoc momento.",
        "Si vinculum aequalitatis",
        "Vicimus cum brachio forti",
        "In sinu tuo, o libertas",
        "Provocavit cor nostrum ipsam mortem!",
        "O patria amata, adorata, salve! salve!",
        "Brasil, somnium intensum, fulgor vividus",
        "Amoris et spei ad terram descendit.",
        "Si futurum est gloria, praeteritum suum",
        "Patria heroica gloriam a conquestu.",
        "Sol libertatis splendet in hoc caelo immenso",
        "Patria amata, salve! salve!"
    ]

class TestHinoTranslation(unittest.TestCase):
    
    def setUp(self):
        self.translated_hino = get_translated_hino()

    
    def test_empty(self):
        
        for line in self.translated_hino:
            self.assertGreater(len(line.strip()), 0, f"A traduo est vazia: {line}")
       
    def test_minimumletter(self):
        for line in self.translated_hino:
            self.assertGreater(len(line.split()), 1, f"A traduo tem menos de 2 palavras: {line}")
    
    def test_especial(self):
        for line in self.translated_hino:
            self.assertTrue(line.isprintable(), f"A traduo contm caracteres especiais: {line}")
  
 #   def test_names_contain_letter_z(self):
 #      
 #       for line in self.translated_hino:
 #           self.assertIn('z', line.lower(), f"A traduo no contm a letra 'z': {line}")
    

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
```
